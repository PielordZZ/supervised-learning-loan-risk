```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'),true_values=['Y','y'],false_values=['N','n'])  #start cleaning by making Y and N's into true/false which are boolean and treatable as 1/0
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'),true_values=['Y','y'],false_values=['N','n'])
```

```python
train_df.columns
```

```python
# Convert categorical data to numeric and separate target feature for training data
train_df = pd.get_dummies(train_df)
train_df.head(10)
```

```python
# Convert categorical data to numeric and separate target feature for training data
test_df = pd.get_dummies(test_df)
test_df.head(10)
```

```python
# add missing dummy variables to testing set
for col in train_df.columns:
    if col not in test_df.columns:
        test_df[col] = 0
test_df = test_df.applymap(lambda x: 0 if (not x) else x )
test_df.head(10)
```

```python

X_test = test_df.drop('loan_status_high_risk', axis=1)
Y_test = test_df['loan_status_high_risk']
for col in X_test.columns:
    if 'loan_status' in col:
        X_test= X_test.drop(col, axis=1)
        test_df = test_df.drop(col, axis=1)

test_df

X_train = train_df.drop('loan_status_high_risk', axis=1)
Y_train = train_df['loan_status_high_risk']
for col in X_train.columns:
    if 'loan_status' in col:
        X_train= X_train.drop(col, axis=1)
        train_df = train_df.drop(col, axis=1)

```

## Create Classification models for the data

Logistic and random forrest models are going to be created to guess if a user has a high risk status of defaulting on a new loan.  The logistic classification is not going to be very accurate without standardizing the data since so many variables are in completely different ranges.  However standardizing the data should not have a large impact on random forest classification since it uses thresholds and binary choices.

Different data will yield different results and the RFC gives a 0.62 accuracy which is significantly better than guessing. which is good for advertizement however using it for something more like actual loan aproval it would probably require a higher accuracy to implement.

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(max_iter =10000)
classifier.fit(X_train,Y_train)
print(f"Training Data Score: {classifier.score(X_train,Y_train)}")
print(f"Testing Data Score: {classifier.score(X_test,Y_test)}")
```

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(random_state=1 ,max_leaf_nodes=3000, n_jobs=64)
classifier.fit(X_train,Y_train)
print(f"Training Data Score: {classifier.score(X_train,Y_train)}")
print(f"Testing Data Score: {classifier.score(X_test,Y_test)}")
```

## Standardizing Data
after doing a non standardized test the data will be standardized using a standard scaler.  This is a mean deviation formula standardizing on a scale of varience.  This allows variables to be compared to each other without issues of scale, such as some variables may be on a scale of 0-1 and others might be in the tens or hundreds of thousands

```python
# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(train_df.drop('loan_status_high_risk', axis=1))
X_train_scaled = scaler.transform(train_df.drop('loan_status_high_risk', axis=1))
X_test_scaled = scaler.transform(test_df.drop('loan_status_high_risk', axis=1))
```

#### models after scaling

after we scale we will see a dramatic increase in accuracy from the logistic model however the random forest will have very similar accuracy.


After applying we see a accuracy of about 0.72 with the logistic which is entering on an accuracy for professional use cases.

THe random forest model does not noticable improve after scaling data. adjusting hyper params would be more useful than scaling data for models like random forest.

```python
# Train the Logistic Regression model on the scaled data and print the model score
classifier = LogisticRegression(max_iter =10000)
classifier.fit(X_train_scaled,train_df['loan_status_high_risk'])
print(f"Training Data Score: {classifier.score(X_train_scaled,train_df['loan_status_high_risk'])}")
print(f"Testing Data Score: {classifier.score(X_test_scaled,test_df['loan_status_high_risk'])}")
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
classifier = RandomForestClassifier(random_state=1 ,max_leaf_nodes=3000, n_jobs=64)
classifier.fit(X_train_scaled,train_df['loan_status_high_risk'])
print(f"Training Data Score: {classifier.score(X_train_scaled,train_df['loan_status_high_risk'])}")
print(f"Testing Data Score: {classifier.score(X_test_scaled,test_df['loan_status_high_risk'])}")
```
